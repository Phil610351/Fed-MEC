```python


import glob
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from os.path import join

sns.set_style("ticks")
sns.set_context("paper", font_scale=1.25) 
```

```python
## Use this if you want the most recent data.
dirs = glob.glob("out/data/*/")
most_recent_dir = sorted(dirs).pop()
# print(most_recent_dir)
Global_G1 = glob.glob(most_recent_dir + "*-mnist-data_Global_G1.csv")
Global_G1_name = sorted(Global_G1).pop()

Final_G1 = glob.glob(most_recent_dir + "*-mnist-Final_LargeDataSetTest_G1.csv")
Final_G1_name = sorted(Final_G1).pop()


Global_G1_data = pd.read_csv(Global_G1_name)
Final_G1_data = pd.read_csv(Final_G1_name)


## Use this if you want the most recent data.
dirs = glob.glob("out/data/*/")
most_recent_dir = sorted(dirs).pop()
# print(most_recent_dir)
Global_MainFL = glob.glob(most_recent_dir + "*-mnist-data_Global_main.csv")
Global_MainFL_name = sorted(Global_MainFL).pop()

Final_MainFL = glob.glob(most_recent_dir + "*-mnist-Final_LargeDataSetTest_MainFL.csv")
Final_MainFL_name = sorted(Final_MainFL).pop()


## Use this if you want the most recent data.
dirs = glob.glob("out/data/*/")
most_recent_dir = sorted(dirs).pop()
# print(most_recent_dir)
Global_ICC = glob.glob(most_recent_dir + "*-mnist-data_Global_ICC.csv")
Global_ICC_name = sorted(Global_ICC).pop()

Final_ICC = glob.glob(most_recent_dir + "*-mnist-Final_LargeDataSetTest_ICC.csv")
Final_ICC_name = sorted(Final_ICC).pop()


Global_ICC_data = pd.read_csv(Global_ICC_name)
Final_ICC_data = pd.read_csv(Final_ICC_name)

Global_MainFL_data = pd.read_csv(Global_MainFL_name)
Final_MainFL_data = pd.read_csv(Final_MainFL_name)


## Use this if you want the most recent data.
dirs = glob.glob("out/data/*/")
most_recent_dir = sorted(dirs).pop()
# print(most_recent_dir)
Global_Muahammed = glob.glob(most_recent_dir + "*-mnist-data_Global_Muhammed.csv")
Global_Muahammed_name = sorted(Global_Muahammed).pop()

Final_Muahammed = glob.glob(most_recent_dir + "*-mnist-Final_LargeDataSetTest_Muhammed.csv")
Final_Muahammed_name = sorted(Final_Muahammed).pop()


Global_Muahammed_data = pd.read_csv(Global_Muahammed_name)
Final_Muahammed_data = pd.read_csv(Final_Muahammed_name)
```
